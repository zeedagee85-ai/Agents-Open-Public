```python
# Install everything you need
!pip install groq gradio -q

import gradio as gr
from groq import Groq
import os

# Get free API key from https://console.groq.com
client = Groq(api_key="YOUR_FREE_GROQ_KEY")  # Replace this

def analyze_sales_call(transcript, focus_area):
    """Your first $500/month agent"""

    prompt = f"""You are an expert sales coach. Analyze this call focusing on: {focus_area}

Call Transcript:
{transcript}

Provide:
1. Score /10
2. Top 3 wins
3. Top 3 improvements
4. Specific phrases to use next time
5. Deal probability %"""

    response = client.chat.completions.create(
        model="llama-3.1-70b-versatile",
        messages=[{"role": "user", "content": prompt}],
        temperature=0.3,
        max_tokens=1000
    )

    return response.choices[0].message.content

# Beautiful UI (clients pay for this)
demo = gr.Interface(
    fn=analyze_sales_call,
    inputs=[
        gr.Textbox(lines=10, label="Paste Sales Call Transcript"),
        gr.Dropdown(
            ["Objection Handling", "Discovery Questions", "Closing", "Rapport Building"],
            label="Focus Area"
        )
    ],
    outputs=gr.Markdown(label="Analysis"),
    title="ğŸ¯ Sales Call Analyzer Pro",
    description="AI-powered sales coaching in 30 seconds"
)

demo.launch(share=True)  # Creates public link instantly
```
